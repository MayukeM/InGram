下面这段代码的作用是：给定一个三元组列表 triplets，根据其中的前两列构建一个有向图 G_ent。具体来说，这段代码的实现步骤如下：

使用 np.array(triplets) 将三元组列表 triplets 转换为一个 NumPy 数组。

使用 [:, 0::2] 对数组进行切片，取出其中的偶数列，即三元组中的第一列和第三列，作为有向图的边列表。

将边列表作为参数，使用 igraph.Graph.TupleList() 函数构建一个有向图 G_ent。该函数的作用是将给定的边列表转换为一个 igraph 图对象。

总之，这段代码的作用是将三元组列表表示的有向图转换为一个 igraph 图对象，便于使用 igraph 库中提供的图算法进行分析和计算。

```python
import igraph
import numpy as np
G_ent = igraph.Graph.TupleList(np.array(triplets)[:, 0::2], directed=True)  # 以实体为节点的图
G_ent.summary()
```

下面这段代码使用了 igraph 库中的 `spanning_tree()` 函数，该函数的作用是在一个图中找到一棵生成树（Spanning Tree）。

生成树是图论中的一个概念，它是指一个无向图的一棵极小的连通子图，它包含了图中所有的顶点，但是没有形成任何的环路。换言之，生成树是一棵包含图中所有顶点的树，它的边数为 $n-1$（其中 $n$ 是图中顶点的数量），并且不包含回路。

在 igraph 库中，`spanning_tree()` 函数可以用于在一个图中找到一棵生成树。它返回一个新的 igraph 图对象，该图对象是原图中的一棵生成树。

因此，这段代码的作用是在变量 `G_ent` 表示的有向图中找到一棵生成树，并将它存储在变量 `spanning` 中。

```python
spanning = G_ent.spanning_tree()  # 生成最小生成树, 作用是将图中的所有节点连接起来，且边的权重之和最小
spanning.summary()
```

```python
G_ent.delete_edges(spanning.get_edgelist())  # 删除最小生成树中的边
G_ent.summary()
```

```python
spanning_list = []
for e in spanning.es:  # spanning.es 返回一个 igraph 中的 EdgeSeq 对象
    e1, e2 = e.tuple  # e.tuple 返回边的两个端点
    e1 = spanning.vs[e1]['name']
    e2 = spanning.vs[e2]['name']
    spanning_list.append((e1, e2))
spanning_list, len(spanning_list)  # 论文中的self.spanning
```

下面这行代码首先将三元组列表 `triplets` 中的每个三元组 $(h,r,t)$ 转换为一个新的三元组 $(t,r+len(id2rel),h)$，然后将这些新的三元组和原来的三元组拼接在一起，得到一个新的三元组列表 `self.triplets_with_inv`，其中包含了每个三元组的逆关系。

具体来说，这行代码的实现步骤如下：

1. 使用列表推导式 `(t, r + len(id2rel), h) for h, r, t in triplets` 将三元组列表 `triplets` 中的每个三元组 $(h,r,t)$ 转换为一个新的三元组 $(t,r+len(id2rel),h)$。

2. 使用 `+` 运算符将上一步得到的新三元组列表和原来的三元组列表 `triplets` 拼接在一起，得到一个新的三元组列表 `[(t, r + len(id2rel), h) for h, r, t in triplets] + triplets`。

3. 使用 `np.array()` 将新的三元组列表转换为一个 NumPy 数组，并将它存储在实例变量 `self.triplets_with_inv` 中。

这样做的目的是为了在知识图谱中加入每个关系的逆关系，以便在查询中考虑到逆关系的影响。例如，如果知识图谱中包含了关系 $(h,r,t)$，那么加入逆关系 $(t,r^{-1},h)$ 后，查询 $(h, r, ?)$ 等价于查询 $(?, r^{-1}, h)$，即查询关系 $(h, r^{-1}, ?)$ 的逆关系。

```python
triplets_with_inv = np.array([(t, r + len(id2rel), h) for h, r, t in triplets] + triplets)
triplets_with_inv, len(triplets_with_inv)
```
